```python
def area(base, altura):
    return base * altura

def vereficaforma(base, altura):
    if base == altura:  # se for quadrado 1, se for retangulo e 0
        shape = "quadrado"
    else:
        shape = "retangulo"
    calculated_area = area(base, altura)
    return f'A figura é um {shape} e sua área é {calculated_area}'


print(vereficaforma(5, 6))
print(vereficaforma(4, 4))
```

```python
### IMC < 18.5 - Magro
### 18.5 < IMC <24.9 - Estotrofico
### 25.0 < IMC < 29.9 - Sobrepeso
### IMC > 30 - Obeso
```

```python
def IMC ( peso, altura):
  return peso / altura**2

def classificaIMC(peso, altura):
  valorIMC = IMC (peso, altura)
  if (valorIMC > 30):
    return 'obeso'
  elif(valorIMC > 25):
    return 'sobrepeso'
  elif(valorIMC > 18.5):
    return 'magro'
```

```python
print(classificaIMC(80, 1.75))
```
